```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
iterations = [*range(1, 52)]
mean = []
for i in iterations:
    with open(f'times/analysis_time_{str(i).zfill(2)}_highways.txt', 'r') as f:
        data = f.readlines()
        data = [float(row.strip()) for row in data]
        mean.append(np.mean(data))

plt.plot(iterations, mean)
plt.title('Nmero de simuladores vs tempo de processamento')
plt.ylabel('Tempo de processamento (ms)')
plt.xlabel('Nmero de simuladores')
plt.show()
```
